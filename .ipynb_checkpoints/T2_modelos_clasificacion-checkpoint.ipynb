````python
def represento_doble_hist(x_1, x_0,  title='', label_1='Clase 1', 
                          label_0='Clase 0'):
    '''
    Función que recibe dos distribuciones de probabilidad y las representa
    en el mismo gráfico
    '''

    #plt.hist(x_1,  density = 0, alpha=0.5, label=label_1, color='red')    
    plt.hist(x_0, density = 0, alpha=0.5, label=label_0, color='green')
    plt.title(title)
    plt.legend(loc='best') 
````

````python
def represento_curva_roc(y, y_prob, title):
    
    fpr, tpr, threshold = roc_curve(y, y_prob[:, 1])
    roc_auc = auc(fpr, tpr)

    plt.title(title)
    plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()
````

````python
def predict_and_get_auc(model):
    
    y_train_prob = model.predict_proba(X_train)
    y_test_prob = model.predict_proba(X_test)

    fpr, tpr, threshold = roc_curve(y_train, y_train_prob[:, 1])
    print("AUC train = ", round(auc(fpr, tpr), 2))

    fpr, tpr, threshold = roc_curve(y_test, y_test_prob[:, 1])
    print("AUC test = ", round(auc(fpr, tpr), 2))
````

### Conversión de categóricas a numéricas y escalado

Como hemos visto que eran categóricas que no siguen un orden, las transformamos a dummies con OneHotEncoder. Las variables numéricas las escalamos. Para hacer ambas transformaciones de una sola vez y poder guardarlas en un fichero .pkl para la posterior productivización, es conveniente utilizar un Pipeline:

````python
# Obtenemos lista con los nombres de las columnas categóricas y numéricas por separado

cat_cols = X.select_dtypes(include = ['object']).columns
num_cols = X.select_dtypes(include = ['integer', 'float']).columns
````

````python
# Definir transformadores para columnas categóricas y numéricas

transformadores = [
    ('onehot', OneHotEncoder(handle_unknown='ignore', drop = "first"), 
     cat_cols),  # OneHotEncoder para las columnas categoricas
    ('scaler', StandardScaler(), num_cols)]  # StandardScaler para las columnas numéricas

# ColumnTransformer facilita aplicar diferentes transformaciones segun la columna
preprocesador = ColumnTransformer(transformadores)

# Crear el pipeline completo con OneHotEncoder seguido de StandardScaler
pipeline = Pipeline(steps=[('preprocesador', preprocesador)])
````

````python
# Ajustar el pipeline sobre el conjunto de datos de train y transformar los datos de train y test

X_train = pipeline.fit_transform(X_train) # el objeto "pipeline" se puede guardar en formato pkl
X_test = pipeline.transform(X_test)
````

**NOTA**: Es conveniente que los transformadores se ajusten sobre los datos de train, aplicándolos luego sobre los datos de test:

- Ajustar (fit) el Scaler y el Encoder solo en el conjunto de entrenamiento.
- Aplicar (transform) las transformaciones en ambos conjuntos (train y test) para que tengan la misma estructura y escala, usando los parámetros aprendidos únicamente del conjunto de entrenamiento.

**Razones por las que conviene ajustar los transformadores solo sobre train:**

1. El ajuste de transformaciones en el conjunto de prueba introduce conocimiento de estos datos en el modelo, que no debería tener acceso a ninguna información del conjunto de prueba durante el entrenamiento. El conjunto de prueba debe representar datos "nuevos" o "no vistos", que sirven para evaluar el rendimiento del modelo en un contexto más realista.

2. Al ajustar el Scaler solo en el conjunto de entrenamiento, estamos asegurando que el modelo aprenda a partir de las estadísticas (media, desviación estándar, etc.) de los datos de entrenamiento. Si usamos ambos conjuntos, el escalado del modelo no reflejará correctamente la distribución de la muestra con la que fue entrenado.

3. En el caso de un Encoder, el conjunto de entrenamiento define la estructura de las categorías. Si nuevas categorías aparecen en el conjunto de prueba, no deben ser consideradas en el entrenamiento (estarías introduciendo una variable con todos los valores a 0 para entrenar).

Si ajustamos el Scaler y el Encoder usando tanto el conjunto de entrenamiento como el de prueba, introducimos un sesgo que hace que la evaluación del modelo sea poco realista: necesitamos asegurarnos de que solo "vea" el conjunto de prueba después del entrenamiento, sin haber utilizado esa información durante el ajuste (ya que los datos de test deben representar un escenario real, donde el modelo será aplicado a datos desconocidos)

### Métricas matriz de confusión

Las probabilidades que devuelve el modelo son valores continuos entre 0 y 1. Para pasarlo a 0s y a 1s (necesario para la matriz de confusión) es necesario usar un umbral de corte. Aquellas observaciones cuya probabilidad predicha sea superior al umbral definido tendrá predicción = 1, en caso contrario predicción = 0.

````python
umbral = 0.5
y_test_pred_log = 1*(y_test_prob_log[:, 1] > umbral)
````

El modelo está muy sobreajustado. Probamos a optimizar los hiperparámetros.

- **Grid search**

Esta técnica permite probar de forma automática distintas combinaciones de los hiperparámetros.

- Indicar rejilla de hiperparámetros que quieres probar
- Para cada posible combinación, la función ajusta el modelo y evalua la performance obtenida. En lugar de evaluar sobre el mismo conjunto con el que entrena, hace cross validation y devuelve el AUC medio
- Devuelve la combinación de hiperparámetros con la que se haya obtenido una mayor performance
- Una vez obtenida la mejor combinación, ajustaremos el modelo sobre nuestro conjunto de train y haremos la evaluación final sobre test.

````python
params = {'max_depth': range(1,10),
         'min_samples_leaf': range(15,30)}

tree = DecisionTreeClassifier()

tree_cv = GridSearchCV(tree, params, cv=3, scoring = 'roc_auc')

tree_cv.fit(X_train,y_train)

tree_cv.best_params_
````

````python
tree =  DecisionTreeClassifier(**tree_cv.best_params_, # max_depth = 4, min_samples_leaf = 23
                            random_state = 0)

tree.fit(X_train, y_train)
````

````python
predict_and_get_auc(tree)
````

## 2.4. Random Forest
---

Como este modelo suele sobreajustar al dejar los hiperparámetros por defecto (al igual que el árbol simple), buscamos los hiperparámetros óptimos:

En este caso, probamos la optimización mediante **RandomizedSearchCV**: en lugar de probar todas las combinaciones de la rejilla, selecciona aleatoriamente x subconjuntos a probar (el nº que se le indique en el parámetro n_iter).

````python
params = {'n_estimators': [100, 200, 300], 
          "min_samples_leaf": range(10,50),
          "max_depth": range(4, 20)}

rf = RandomForestClassifier()

rf_random = RandomizedSearchCV(rf, params, cv=3, n_iter = 20, random_state = 0)

rf_random.fit(X_train,y_train)

rf_random.best_params_
````

````python
rf_random.cv_results_['params']
````

````python
rf =  RandomForestClassifier(**rf_random.best_params_,
                             random_state = 0)

rf.fit(X_train, y_train)
````

````python
predict_and_get_auc(rf)
````

De esta forma, para obtener predicciones futuras solamente tendrás que cargar el modelo y los transformadores (sin necesidad de volver a entrenar) y pasarle al `.predict` los nuevos datos pre-procesados.
